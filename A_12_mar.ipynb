```python
"""Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results."""
```

```python
import scipy.stats as stat
sample_mean = 50
sample_std = 5
n =100
z_critical = stat.norm.ppf(0.975)
margin_of_error = z_critical * (sample_std / (n ** 0.5))

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print("The 95% confidence interval is:", confidence_interval)
```

```python
"""Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05."""
```

```python
import numpy as np
from scipy.stats import chisquare
observed_freq = np.array([24, 16, 19, 12, 9, 20])
expected_freq = np.array([.2, .2, .2, .1, .1, .2]) * sum(observed_freq)
test_stat, p_value = chisquare(observed_freq,f_exp = expected_freq)

if p_value < 0.05:
    print("The null hypothesis is rejected.")
else:
    print("The null hypothesis is not rejected.")

# printing the test statistics and p-value
print("The test statistic is:", test_stat)
print("The p-value is:", p_value)
```

```python
"""Q3. Use Python to calculate the chi-square statistic and p-value for a contingency table with the following
data:"""
```

```python
from scipy.stats import chi2_contingency

# creating the contingency table
contingency_table = np.array([[20,15], [10,25], [15,20]])

# calculating the chi-square test statistic, p-value, degrees of freedom, and expected frequencies
chi2_stat, p_val, dof, expected_freq = chi2_contingency(contingency_table)

# printing the results
print("The chi-square test statistic is:", chi2_stat)
print("The p-value is:", p_val)
print("The degrees of freedom is:", dof)
print("The expected frequencies are:\n", expected_freq)
```

```python
"""Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results."""
```

```python
import numpy as np
import scipy.stats as stats
mean =75
std_dev =12
n=100
std_err = std_dev /np.sqrt(n)
t_value = stats.t.ppf(1-0.1/2,n-1)
margin_of_error = t_value *std_err

conf_int = (mean-margin_of_error, mean+margin_of_error)
print("The 90% confidence interval for the sample is:", conf_int)
```

```python
"""Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15."""
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# defining the degrees of freedom
df = 10

# defining the range of x values for the plot
x = np.linspace(0, 30, 500)

# creating the plot
fig, ax = plt.subplots(figsize=(8, 5))

# plotting the chi-square distribution
ax.plot(x, stats.chi2.pdf(x, df), 'b-', lw=2, alpha=0.6, label='chi-square pdf')

# shading the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 500)
ax.fill_between(x_shade, 0, stats.chi2.pdf(x_shade, df), alpha=0.5, color='g', label='p-value')

# adding labels and a legend to the plot
ax.set_xlabel('Chi-square statistic')
ax.set_ylabel('Probability density')
ax.set_title('Chi-square distribution with {} degrees of freedom'.format(df))
ax.legend(loc='best')

# displaying the plot
plt.show()
```

```python
"""Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke."""
```

```python
import numpy as np
import scipy.stats as stats

# defining the sample statistics
n = 1000
p = 520/n

# calculating the standard error of the proportion
std_error = np.sqrt((p * (1 - p)) / n)

# calculating the margin of error using the z-distribution
z_value = stats.norm.ppf(1 - 0.01 / 2)
margin_of_error = z_value * std_error

# calculating the confidence interval
conf_int = (p - margin_of_error, p + margin_of_error)

# printing the confidence interval
print("The 99% confidence interval for the true proportion of people who prefer Coke is:", conf_int)
```

```python
"""Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05."""
```

```python
import numpy as np
import scipy.stats as stats

# defining the observed frequencies
obs_freq = np.array([55, 45])

# defining the expected frequencies under the null hypothesis of a fair coin
exp_freq = np.array([50, 50])

# calculating the chi-square statistic and p-value
chi2_stat, p_value = stats.chisquare(obs_freq, exp_freq)

# defining the significance level
alpha = 0.05

# printing the results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("The p-value is less than the significance level of", alpha)
    print("Reject the null hypothesis of a fair coin.")
else:
    print("The p-value is greater than or equal to the significance level of", alpha)
    print("Fail to reject the null hypothesis of a fair coin.")
```

```python
"""Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05."""
```

```python
import numpy as np
import scipy.stats as stats

# defining the sample size, sample mean, sample standard deviation, and null hypothesis value
n = 30
sample_mean = 72
sample_std_dev = 10
null_mean = 70

# calculating the t-statistic and p-value assuming a two-tailed test
t_stat, p_value = stats.ttest_1samp(np.random.normal(sample_mean, sample_std_dev, n), null_mean)

# defining the significance level
alpha = 0.05

# printing the results
print("t-statistic:", t_stat)
print("p-value:", p_value)
if p_value < alpha:
    print("The p-value is less than the significance level of", alpha)
    print("Reject the null hypothesis.")
else:
    print("The p-value is greater than or equal to the significance level of", alpha)
    print("Fail to reject the null hypothesis.")
```
