```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure, output_file, show
import numpy as np

np.random.seed(42)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

fig = figure(title="Scatter plot example", x_axis_label="x", y_axis_label="y")

fig.circle(x, y, size=10, color="navy", alpha=0.5)

output_file("scatter.html")
show(fig)
```

```python
"""Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example."""

"""Glyphs in Bokeh are the visual marks or symbols that represent data points on a plot. Bokeh provides a variety of glyph types, including circles, squares, lines, and more."""
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

source = ColumnDataSource(data=dict(x=x, y=y))

p = figure(plot_width=400, plot_height=400)

p.circle(x='x', y='y', source=source, size=10, color='navy')

output_file('glyphs.html')

show(p)
```

```python
"""A Bokeh server is a Python process that allows you to create and serve interactive Bokeh plots and apps that can be updated in real-time. With a Bokeh server, you can create 
dynamic and responsive data visualizations that can handle user interactions and respond to changes in the underlying data.

To use the Bokeh server, you first need to define your plot or app as a Python function or class that returns a Bokeh plot or layout. You can then start the Bokeh server and point 
it to your function or class. Once the server is running, you can access your plot or app in a web browser and interact with it."""
```
