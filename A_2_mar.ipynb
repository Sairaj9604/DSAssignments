```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot."""
```

```python
"""A scatter plot is a type of plot that displays the relationship between two variables as a collection of points in a 2D coordinate system.
Each point on the plot represents a unique combination of values for the two variables being plotted."""
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axix")
plt.title("Scatterplot")
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function."""
```

```python
"""The subplot() function is used to create multiple plots within the same figure. It allows users to divide the figure into a grid of 
subplots and then plot different data or variations of the same data in each subplot."""
```

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

fig.suptitle('Four Line Plots')

plt.show()
```

```python
"""Q5 What is a box plot? Why is it used? Using the following data plot a box plot."""
```

```python
"""A box plot, also known as a box-and-whisker plot, is a type of statistical chart used to represent a set of data through its quartiles. 
The plot displays the distribution of the data based on five key statistics: minimum value, maximum value, median, first quartile (Q1), and 
third quartile (Q3)."""
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
labels = ['Box 1', 'Box 2']

fig, ax = plt.subplots()
ax.boxplot(data, labels=labels)

plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```
